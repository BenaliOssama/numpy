```python
import numpy as np
data = np.loadtxt('data/model_forecasts.txt', delimiter=',')
print(data.shape)
print(data)
```

```python
import itertools
```

```python
teams = np.arange(10)
print(teams)
```

```python
permutations = np.array(list(itertools.permutations(teams)))
print(permutations)
print(permutations.shape)
```

```python
team1 = permutations[:, 0::2]  # columns 0,2,4,6,8
team2 = permutations[:, 1::2]  # columns 1,3,5,7,9

print(team1.shape)
print(team2.shape)
```

```python
differences = data[team1, team2]
print(differences)
print(differences.shape)
```

```python
scores = np.sum(differences ** 2, axis=1)
print(scores)
print(scores.shape)
```

```python
best_index = np.argmin(scores) # index of the smallest value
best_pairing = permutations[best_index]  # the smallest value itself
print(best_pairing)
```

```python
result = best_pairing.reshape(2, 5, order='F') # fortrone
print(result)
```
