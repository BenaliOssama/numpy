```python
import numpy as np
```

```python
data = np.genfromtxt('data/winequality-red.csv', 
                      delimiter=';', 
                      skip_header=True, 
                      dtype=np.float32)
print(data.nbytes)  # should print 76800
print(data.shape)   # should be (1599, 12)
```

```python
rows = data[[1, 6, 11], :]
print(rows)
clean_rows = rows[~np.isnan(rows).any(axis=1)]
print(clean_rows)
#print(rows[~np.isnan(rows)#true for any nan
 #   .any(axis=1)#if any is true true
  #  ])#~ reverse boolean -> drop any row with missing value
```

```python
print(np.any(data[:, 11] > 20))
```

```python
print(np.nanmean(data[:, 11]))
```

```python
ph = data[:, 8]
print('min:   ', np.nanmin(ph))
print('max:   ', np.nanmax(ph))
print('25th:  ', np.nanpercentile(ph, 25))
print('50th:  ', np.nanpercentile(ph, 50))
print('75th:  ', np.nanpercentile(ph, 75))
print('mean:  ', np.nanmean(ph))
```

```python
sulphate_20th = np.nanpercentile(data[:, 9], 20)
mask = data[:, 9] < sulphate_20th
print(mask)
print(np.nanmean(data[mask, 11]))
```

```python
best_quality = np.nanmax(data[:, 11])
worst_quality = np.nanmin(data[:, 11])

best_mask = data[:, 11] == best_quality
worst_mask = data[:, 11] == worst_quality

print(np.nanmean(data[best_mask], axis=0))
print(np.nanmean(data[worst_mask], axis=0))
```
